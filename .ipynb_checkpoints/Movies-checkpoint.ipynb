```python
wiki_movies = [movie for movie in wiki_movies_raw
              if ('Director' in movie or 'Directed by' in movie) # want to have movies that have a director
                 and 'imdb_link' in movie #or directed by and have an imdb link
                 and 'No. of episodes' not in movie] #we want movies not tv shows
```

```python
def clean_movie(movie):
    movie = dict(movie) # create a nondestructive copy 
    alt_titles = {}
    for key in {'Also known as', 'Arabic', 'Cantonese', 'Chinese','French','Hangul','Hebew','Hepburn',
                'Japanese','Literally','Mandarin','McCune-Reischauer','Original title','Polish',
                'Revised Romanization','Romanized','Russian','Simplified','Traditional','Yiddish'}:
        if key in movie:
                alt_titles[key] = movie[key]
                movie.pop(key)
    if len(alt_titles) > 0:
            movie['alt_titles'] = alt_titles
            
    return movie
```

```python
# Create a function to help find and delete unneeded data
clean_movies = [clean_movie(movie) for movie in wiki_movies]
```

```python
wiki_movies_df = pd.DataFrame(clean_movies)
sorted(wiki_movies_df.columns.tolist())
```
